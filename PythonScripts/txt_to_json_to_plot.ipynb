```python
import json 
from matplotlib import pyplot as plt
import os
import numpy as np
import re

os.chdir("../log_files")

batch_list = []


with open('toplvl_some_parents_rec.log', 'r') as file_in:
    
    data = file_in.read().replace('\n', '')

#header, 5,10,20,50,100,200
batch_list = data.split('json_conf:')
del batch_list[0]

#del batch_list[1]


def del_duplicates(lod):
    
    seen = set()
    added_seen = seen.add
    return [x for x in lod if not (x in seen or added_seen(x))]


loss_row = []
mapr_row = []
meanr_row = []

for index, val in enumerate(batch_list):
    
#     print(val)
#     break

    losses = re.findall('"loss":(.+?), ', val)
    losses = [float(x) for x in losses]
    
    mapr = [float(x) for x in re.findall('"map_rank":(.+?), ', val)]
    meanr = [float(x) for x in re.findall('"mean_rank":(.+?), ', val)]
    
    set_meanr= set(meanr)
    
    mapr = del_duplicates(mapr)
    meanr = del_duplicates(meanr)
    losses = del_duplicates(losses)
    
    if len(losses) != len(mapr) or len(losses) != len(meanr):
        print('Different lengths in iter {}. losses: {}, mapr: {}, meanr: {}'.format(index,len(losses),len(mapr),len(meanr)))
        #print(meanr)
        #print(set_meanr, len(set_meanr))
    
#     print(len(losses))
#     print(len(mapr))
#     print(len(meanr))
#     print(losses)
#     break
    
    loss_row.append(losses)
    mapr_row.append(mapr)
    meanr_row.append(meanr)
    



epochs = range(len(losses))


dims=[5,10,20,50,100,200]

#TODO NICER SUBPLOTS OF ALL OF THEM SIDE BY SIDE

fig, axs = plt.subplots(6,2,figsize=(30,25))


for index, row in enumerate(loss_row):

    
    
    title1 = 'Loss and Map for Dim {}'.format(dims[index])
    axs[index, 0].set_title(title1)
    axs[index, 0].plot(epochs,loss_row[index],'r+', label='loss_row[index] = loss')
    axs[index, 0].plot(epochs,mapr_row[index],'bo', label='mapr_row[index] = MAP rank')

    
    title2 = 'Mean rank for Dim {}'.format(dims[index])
    axs[index, 1].set_title(title2)
    axs[index, 1].plot(epochs,meanr_row[index],'gs', label='meanr_row[index] = mean rank')


    
#     axs1.set_title('Loss and MAP')
#     axs1.plot(epochs,loss_row[index],'r+', label='loss_row[index] = loss')
#     axs1.plot(epochs,mapr_row[index],'bo', label='mapr_row[index] = MAP rank')
    
#     axs2.set_title('Mean rank')
#     axs2.plot(epochs,meanr_row[index],'gs', label='meanr_row[index] = mean rank')
    
    axs[index,0].legend()
    axs[index,1].legend()
    #plt.show()
    
    for ax in axs.flat:
        ax.set(xlabel='epochs', ylabel='metrics')
        ax.label_outer()

        
os.chdir("../plots")
file_name = 'Metrics Plot for'+ 'toplvl_some_parents'
plt.savefig(file_name+'.png')
    

    
    
            

        
```
