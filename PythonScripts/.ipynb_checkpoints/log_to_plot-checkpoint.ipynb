```python
import json 
from matplotlib import pyplot as plt
import os
import numpy as np
import re

os.chdir("../log_files")

batch_list = []

file = "notrace_chrystall.log"
name ,log = file.split('.log')


with open(file, 'r') as file_in:
    
    data = file_in.read().replace('\n', '')

#header, 5,10,20,50,100,200
batch_list = data.split('}json_stats: ')

del batch_list[0:1]



def del_duplicates(lod):
    
    seen = set()
    added_seen = seen.add
    return [x for x in lod if not (x in seen or added_seen(x))]


loss_row = []
mapr_row = []
meanr_row = []

for index, val in enumerate(batch_list):
    
#     print(val)
#     break

    losses = re.findall('"loss":(.+?), ', val)
    losses = [float(x) for x in losses]

    
    mapr = [float(x) for x in re.findall('"map_rank":(.+?), ', val)]
    meanr = [float(x) for x in re.findall('"mean_rank":(.+?), ', val)]
    
    set_meanr= set(meanr)
    
    mapr = del_duplicates(mapr)
    meanr = del_duplicates(meanr)
    losses = del_duplicates(losses)

    
    if len(losses) != len(mapr) or len(losses) != len(meanr):
        print('Different lengths in iter {}. losses: {}, mapr: {}, meanr: {}'.format(index,len(losses),len(mapr),len(meanr)))
        print(meanr)
        print(set_meanr, len(set_meanr))
    

    
    loss_row.extend(losses)
    mapr_row.extend(mapr)
    meanr_row.extend(meanr)
    



epochs = range(len(loss_row))


#TODO NICER SUBPLOTS OF ALL OF THEM SIDE BY SIDE

fig, axs = plt.subplots(1,2,figsize=(35,10))


    
axs[0].set_title('Loss and MAP')
axs[0].plot(epochs,loss_row,'r+', label='loss_row = loss')
axs[0].plot(epochs,mapr_row,'bo', label='mapr_row = MAP rank')
axs[0].set_xlabel(xlabel='epochs')

axs[1].set_title('Mean rank')
axs[1].plot(epochs,meanr_row,'gs', label='meanr_row = mean rank')
axs[1].set_xlabel(xlabel='epochs')
    
axs[0].legend()
axs[1].legend()



plt.show()



        
os.chdir("../plots")
file_name = 'Metrics Plot '+ name
plt.savefig(file_name+'.svg')
    

    
    
            

        
```
